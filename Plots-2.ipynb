```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y)
plt.title('sin function')
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
figure=plt.figure()
ax=plt.axes()
ax.plot(x,y,'c-*',label='sin(x)')
ax.plot(x,y+1,'r-',label='sin(x)+1')
#ax.set_title('sin function')
#ax.set_xlim(-1,11)
#ax.set_ylim(-2.5,2.5)
ax.set(title='sin function', xlim=(-1,11), ylim=(-2.5,2.5), xlabel='x')
ax.legend()
```

```python
x=np.linspace(1,10,1000)
fig,axs=plt.subplots(2)
axs[0].plot(x, np.sin(x), color='r')
axs[1].scatter(x, np.cos(x), color='black', marker='+')
```

```python
x=np.linspace(0,np.pi,10)
y=np.sin(x)
fig,axs=plt.subplots(2,2)

axs[0,0].plot(x, y,color='red')
axs[0,1].plot(x,-1*y,color='blue')
axs[1,0].plot(x,np.sin(x), color='green')
axs[1,1].scatter(x, -np.sin(x),color='black',marker='o')
```
